```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
product_names = []
prices = []
specifications = []
availability_list = []

# Function to scrape details from each product link
def scrape_product_details(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract Product Name
    name = soup.find('h1', class_='sc-de5a1c50-17 dJnofv').text.strip() if soup.find('h1', class_='sc-de5a1c50-17 dJnofv') else 'N/A'
    product_names.append(name)

    # Extract Price
    price = soup.find('div', class_='priceNow').text.strip() if soup.find('div', class_='priceNow') else 'N/A'
    prices.append(price)

    # Extract Specifications (if available)
    specs = soup.find('div', class_='sc-kfGgVZ').text.strip() if soup.find('div', class_='sc-kfGgVZ') else 'N/A'
    specifications.append(specs)

    # Extract Availability (if available)
    availability = 'Available' if soup.find('button', class_='sc-qGWeR') else 'Out of stock'
    availability_list.append(availability)
```

```python
class Solution:
    def reverse(self, x: int) -> int:
        # Define the 32-bit integer range limits
        INT_MAX = 2**31 - 1
        INT_MIN = -2**31

        # Handle the sign
        sign = -1 if x < 0 else 1
        x *= sign

        # Reverse the integer by converting it to a string, reversing the string, and converting back to integer
        reversed_x = int(str(x)[::-1])

        # Restore the original sign
        reversed_x *= sign

        # Check for overflow
        if reversed_x < INT_MIN or reversed_x > INT_MAX:
            return 0

        return reversed_x
```

```python

import requests
from bs4 import BeautifulSoup

# URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/List_of_continents_by_GDP"

# Send a GET request to the webpage
response = requests.get(url)

# Parse the webpage content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table containing GDP data
table = soup.find('table', {'class': 'wikitable'})

# Extract the header
headers = [header.text.strip() for header in table.find_all('th')]

# Extract the rows from the table
rows = []
for row in table.find_all('tr')[1:]:  # skip the header row
    cells = row.find_all(['td', 'th'])
    rows.append([cell.text.strip() for cell in cells])

# Display the results
print(headers)
for row in rows:
    print(row)
```
